# 现在数据不符合deepAR模型的数据输入格式，因此我们需要做以下工作：
- 1、选择日期特征和销量特征；
- 2、把选择出来的特征按：行为日期，列为商品进行排列

## 1、选择日期特征和销量特征:

```python
all_data = pd.concat([train_data, test_data]) # 把训练集和测试集按列拼接，好处是便于处理
all_data
```

```python
sale_data = all_data.loc[:, ["sale_ord_dt", "item_sku_id", "rev_addr_province_id", "sale_qtty"]]
sale_data.set_index(["sale_ord_dt"], inplace=True) # 指定sale_ord_dt作为索引
sale_data.index = pd.to_datetime(sale_data.index) # 让行索引转换成datetime类型
```

```python
sale_data
```

## 2、把选择出来的特征按：行为日期，列为商品进行排列

### 看一下一共有多少个商品

```python
# all_data = pd.DataFrame() # 创建一个空df表
i = 0
for name, group in sale_data.groupby(["item_sku_id"]):
#     all_data[str(name)] = group["sale_qtty"] # 把每一列添加到后面
    i += 1

# all_data.head()
i # 1134个商品
```

### 按item_sku_id划分为组，然后把同一个商品不同地区同一天的销量加起来，按行索引（日期）拼接

```python
all_sale_data = pd.DataFrame() # 创建一个空df表

for name, group in sale_data.groupby(["item_sku_id"]):
    group = group.sort_values(by="sale_ord_dt", ascending=True) # 把group按照index从小到大排序
#     print(name, group)
    summary = group.groupby(["sale_ord_dt"]).sum() # 把分的这个组，再按照日期分组,然后把相同日期的销量加起来
#     print(summary)
    all_sale_data[str(int(name))] = summary["sale_qtty"] # 把每一列添加到后面
#     break
```

# 缺失值处理
- 1、缺太多的列，就直接删除
- 2、缺失少量的，使用均值填充

```python
import numpy as np

for column in all_sale_data.columns: # 按列遍历
    nan_nums = all_sale_data[column].isnull().sum() # 统计每一列缺失值的个数
    if nan_nums < len(all_sale_data[column]) * 0.75: # 缺失值的个数小于总数的75%，则填充
        if np.all(pd.notnull(all_sale_data[column])) == False:
            all_sale_data[column].fillna(0, inplace=True) # 使用0填充
    else: # 缺失值的个数大于等于总数的75%，则删除这一列
        all_sale_data.drop(labels=column, axis=1, inplace=True)
```
